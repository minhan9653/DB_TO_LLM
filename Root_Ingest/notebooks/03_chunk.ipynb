# 03. Chunk Test
이 파일은 파싱된 문서를 단순 규칙으로 청킹하는 모듈입니다.

```python
from pathlib import Path
import sys

WORKDIR = Path.cwd()
REPO_ROOT = WORKDIR.parent if WORKDIR.name == "notebooks" else WORKDIR
if REPO_ROOT.name == "Root_Ingest":
    REPO_ROOT = REPO_ROOT.parent
PROJECT_ROOT = REPO_ROOT / "Root_Ingest"
for path_item in (REPO_ROOT, PROJECT_ROOT):
    if str(path_item) not in sys.path:
        sys.path.append(str(path_item))
```

```python
from Root_Ingest.utils.config_loader import load_config
from Root_Ingest.utils.path_utils import resolve_path
from Root_Ingest.ingest.parser_service import load_parsed_documents
from Root_Ingest.ingest.chunk_service import chunk_documents

config = load_config(PROJECT_ROOT / "config" / "config.yaml")
parsed_dir = resolve_path(config["paths"]["parsed_dir"], PROJECT_ROOT)
chunks_dir = resolve_path(config["paths"]["chunks_dir"], PROJECT_ROOT)

parsed_docs = load_parsed_documents(parsed_dir / "parsed_documents.jsonl")
chunks = chunk_documents(
    parsed_documents=parsed_docs,
    chunk_size=int(config["chunking"]["chunk_size"]),
    chunk_overlap=int(config["chunking"]["chunk_overlap"]),
    output_path=chunks_dir / "chunks.jsonl",
)
print(f"Chunks: {len(chunks)}")
chunks[0].chunk_text[:300] if chunks else "No chunk output"
```
