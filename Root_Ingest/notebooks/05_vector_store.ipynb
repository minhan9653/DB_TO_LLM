# 05. Vector Store Test
이 파일은 임베딩 데이터를 ChromaDB에 저장하는 단계입니다.

```python
from pathlib import Path
import sys

WORKDIR = Path.cwd()
REPO_ROOT = WORKDIR.parent if WORKDIR.name == "notebooks" else WORKDIR
if REPO_ROOT.name == "Root_Ingest":
    REPO_ROOT = REPO_ROOT.parent
PROJECT_ROOT = REPO_ROOT / "Root_Ingest"
for path_item in (REPO_ROOT, PROJECT_ROOT):
    if str(path_item) not in sys.path:
        sys.path.append(str(path_item))
```

```python
from Root_Ingest.utils.config_loader import load_config
from Root_Ingest.utils.path_utils import resolve_path
from Root_Ingest.ingest.embedding_service import load_embeddings
from Root_Ingest.ingest.vector_store_service import upsert_embeddings_to_chroma, sample_from_collection

config = load_config(PROJECT_ROOT / "config" / "config.yaml")
embeddings_dir = resolve_path(config["paths"]["embeddings_dir"], PROJECT_ROOT)
chroma_dir = resolve_path(config["paths"]["chroma_dir"], PROJECT_ROOT)

embeddings = load_embeddings(embeddings_dir / "embeddings.jsonl")
stored_count = upsert_embeddings_to_chroma(
    embeddings=embeddings,
    persist_directory=chroma_dir,
    collection_name=config["vector_store"]["collection_name"],
    batch_size=int(config["vector_store"]["batch_size"]),
)
sample = sample_from_collection(
    persist_directory=chroma_dir,
    collection_name=config["vector_store"]["collection_name"],
    limit=3,
)
print(f"Stored vectors: {stored_count}")
sample
```
